# **Applying Annova Test to Accept or Reject The Hypothesis**

```python
import numpy as np
import scipy.stats as stats


# Perform Welch's ANOVA
#calculate f_satistic
f_statistic, p_value = stats.f_oneway(sindh, punjab, balochistan,GB,kpk,other)

k=6 #total groups
n=303 #total rows
df_between = k-1
df_within = n-k
alpha = 0.05
# Calculate critical F-value
critical_f_value = stats.f.ppf(1 - alpha, df_between, df_within)

# Display results
print("Welch's F-Statistic:", f_statistic)
print("F-Cirtical value: ",critical_f_value )
print("P-Value:", p_value)
# Compare with the obtained F-statistic
if f_statistic < critical_f_value:
    print("There is no significant Difference in satisfaction ratings among individuals from different geographical regions..")
else:
    print("There is  significant Difference in satisfaction ratings among individuals from different geographical regions..")

```
